```python
import pandas as pd
import ast
from transformers import pipeline

#returns positivity from 0 to 1, 0 is negative and 1 is positive
def classify(txt):
    sent_obj = classifier(txt)[0]
    if sent_obj["label"] == "NEGATIVE":
        return(1-sent_obj["score"])
    else:
        return(sent_obj["score"])
classifier = pipeline("sentiment-analysis")

df_hf = pd.read_csv("processed_new_vid.csv")


for i in range(602):
    sent_vals = []
    print(f"Cycle: {i+1}")
    comment_ls = ast.literal_eval(df_hf.iloc[i,6])
    for comment in comment_ls:
        if comment.strip() != "":
            sent_vals.append(classify(comment))
    sent_avg = sum(sent_vals)/len(sent_vals)
    df_hf.iloc[i,6] = sent_avg

df_hf.to_csv("process_digitize1.csv", index=False)
```
